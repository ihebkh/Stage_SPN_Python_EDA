```python
import matplotlib.pyplot as plt
import seaborn as sns

# Compter les occurrences de chaque marque
brand_counts = df['model'].value_counts()

# Histogramme des marques de voitures
plt.figure(figsize=(10, 6))
sns.barplot(x=brand_counts.index, y=brand_counts.values, palette='viridis')
plt.xticks(rotation=90)
plt.title('Distribution des marques de voitures')
plt.xlabel('Marque')
plt.ylabel('Nombre')
plt.show()
```

```python
types_energie = df['energy'].unique()

# Créer des pie charts pour chaque type d'énergie
for energy_type in types_energie:
    # Filtrer les données pour le type d'énergie actuel
    filtered_data = df[df['energy'] == energy_type]
    
    # Compter les occurrences de chaque marque pour ce type d'énergie
    brand_counts = filtered_data['model'].value_counts()
    
    # Créer le pie chart
    plt.figure(figsize=(5, 5))
    brand_counts.plot.pie(autopct='%1.1f%%', colors=sns.color_palette('viridis', len(brand_counts)))
    plt.title(f'Répartition des marques pour l\'énergie: {energy_type}')
    plt.ylabel('')
    plt.show()
```

```python
tiers = df['tier'].unique()
for tier in tiers:
    filtered_data = df[df['tier'] == tier]
    model_counts = filtered_data['model'].value_counts()
    plt.figure(figsize=(5, 5))
    model_counts.plot.pie(autopct='%1.1f%%', colors=sns.color_palette('viridis', len(model_counts)))
    plt.title(f'Répartition des modèles pour le niveau: {tier}')
    plt.ylabel('')
    plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='tier', hue='isActive', data=df, palette='viridis')
plt.title('Nombre de voitures actives par niveau (tier)')
plt.xlabel('Niveau (tier)')
plt.ylabel('Nombre de voitures')
plt.legend(title='Active', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
df['year'] = pd.to_numeric(df['year'], errors='coerce')
plt.figure(figsize=(14, 8))
sns.countplot(x='year', hue='tier', data=df, palette='viridis')
plt.xticks(rotation=90)
plt.title('Répartition des voitures par année de fabrication et par niveau (tier)')
plt.xlabel('Année de fabrication')
plt.ylabel('Nombre de voitures')
plt.legend(title='Niveau (tier)', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='model', y='passengerCount', data=df, palette='viridis', s=100)
plt.xticks(rotation=90)
plt.title('capacite de voiture')
plt.xlabel('Marque')
plt.ylabel('Nombre de passagers')
plt.show()
```
